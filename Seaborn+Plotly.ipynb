```python
#Displot
tips_data=sns.load_dataset("tips")
sns.displot(data=tips_data,x="total_bill",kde=True,color ='m',alpha=0.1,bins=15)
```

```python
#Kde Plot
tips_data=sns.load_dataset("tips")
sns.kdeplot(data=tips_data,x="total_bill",color ='orange',alpha=0.7,hue="sex")
```

```python
#Fill with color
sns.kdeplot(data=tips_data,x="total_bill",color ='m',alpha=0.1,fill=True,hue="sex")
```

###Pair Plot and Joint Plot using Seaborn

```python
Student.tail(5)
```

```python
#Pair Plot
Student_info=Student[["Marks","study_hours","test_score","gender","attendance_rate"]]
sns.pairplot(data=Student_info,hue="gender")
```

```python
#kind='scatter', 'kde', 'hist', 'reg'
sns.pairplot(data=Student_info,hue="gender",kind='reg')
```

```python
#kind='scatter', 'kde', 'hist', 'hex', 'reg', 'resid'
sns.jointplot(data=Student_info,x="study_hours",y="test_score",hue="gender")
```

```python
#kind='scatter', 'kde', 'hist', 'hex', 'reg', 'resid'
sns.jointplot(data=Student_info,x="study_hours",y="test_score",hue="gender",kind='kde')#Use of `hue` with `kind='hex','reg','resid'` are not currently supported.
```

#Plotly
### Plotly Python library for creating interactive data visualizations

###Scatter Plot with Plotly

```python
import plotly.express as px
fig=px.scatter(Student,x="study_hours",y="test_score",color="gender")
fig.show()
```

```python
#hover_data add extra info
fig=px.scatter(Student,x="study_hours",y="test_score",color="gender",hover_data=["Tshirt_size","hostel","subject"])
fig.show()
```

```python
#size respect to T-Shirt
fig=px.scatter(Student,x="study_hours",y="test_score",color="gender",size="Tshirt_size")
fig.show()
```

```python
#Add Animation Frame
fig=px.scatter(Student,x="study_hours",y="test_score",color="gender",animation_frame="class_level")
fig.show()
```

### Line Plot using Plotly

```python
Student.head(3)
```

```python
fig=px.line(Student,x="hostel",y="week",color="gender")
fig.show()
```
