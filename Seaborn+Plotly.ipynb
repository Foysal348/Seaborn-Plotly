# Facet Plot with Seaborn

```python
Student=pd.read_csv("/content/gdrive/MyDrive/Copy of student_dataset_complete.csv")
Student.head(5)
```

```python
sns.scatterplot(x="study_hours", y="test_score", data=Student,hue="gender",style="class_level")
```

```python
sns.relplot(kind="scatter",x="study_hours", y="test_score", data=Student,hue="gender",style="class_level")
```

```python
sns.relplot(kind="scatter",x="study_hours", y="test_score", data=Student,hue="gender",style="class_level",col="gender")
```

```python
sns.relplot(kind="scatter",x="study_hours", y="test_score", data=Student,hue="gender",style="class_level",col="gender",row="hostel")
```

```python
sns.relplot(kind="scatter",x="study_hours", y="test_score", data=Student,hue="gender",style="class_level",col="week",row="hostel")
```

```python
sns.relplot(kind="scatter",x="study_hours", y="test_score", data=Student,hue="gender",style="class_level",col="week",col_wrap=2)
```
